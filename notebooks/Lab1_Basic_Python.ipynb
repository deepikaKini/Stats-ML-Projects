2.3 Check if a string variable contains a specified substring. Print "It is true" if it does or print "It is not".

Expected input: x="I don't like this song", substring="music",

expected result: It is not.

You can use either find(), i.e., fullstring.find(substring) returns the index of the first occurence, will return -1 if the substring is not in the fullstring.

You can also use: in, i.e., substring in fullstring. It returns True or False

In case you need the syntax for if-else, here is the link: https://www.w3schools.com/python/python_conditions.asp

```
x = "I don't like this song"
substring = "music"
a = substring in x
if a == True:
  print("It is true")
elif a == False:
  print("It is not")
```
