13. Resume the dataset to the origninal one.

Replace the null value of Cabin using its mode, replace the null value of Age using its mean

print the table after the replace

```python
TitanicDataCopy = TitanicData.copy()
Cabinmode = TitanicDataCopy['Cabin'].mode()
Agemean = TitanicDataCopy['Age'].mean()
TitanicDataCopy["Cabin"].fillna(Cabinmode[1],inplace = True)
TitanicDataCopy["Age"].fillna(Agemean,inplace = True)
TitanicDataCopy
```

# Outliers:
15. Define an acceptable range for "Age" and remove the rows falling out of the range ( you can have your own defintion of the normal range)

```python
TitanicDataCopy = TitanicData.copy()
Agestd = TitanicDataCopy['Age'].std()
lowerlimit = Agemean - Agestd
upperlimit = Agemean + Agestd
TitanicDataCopy = TitanicDataCopy.drop(TitanicDataCopy[(TitanicDataCopy.Age<lowerlimit)|(TitanicDataCopy.Age>upperlimit)].index)
TitanicDataCopy
```
