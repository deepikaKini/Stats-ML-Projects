4. Use box plots to show the age value distribution for each Embarked value

```
Titanic_data.boxplot(column='Age', by='Embarked')
```

5. Use bar chart to plot the number of passengers for survived and perished separately. 

  1. Use red colors for the bars. 
  2. label the bars as "Survived" and "Not Survived"

Hints: 

  1. First compute the number of passengers for each class and generate a series of it. 
  2. Use Series.to_frame() to convert the type to data frame. https://pandas.pydata.org/docs/reference/api/pandas.Series.to_frame.html. So you can use DataFrame.plot.bar() to plot the bar chart. 
  3. Now the dataframe is (0,1). Change the index of the dataframe to ('Survived', 'Not Survived') to correctly label each bar. 
  4. Set the color as 'red' in dataframe.plot.bar() function. 

```
bar_mf=Titanic_data.groupby('Survived').sum().agg('PassengerId').to_frame()
bar_mf.index=('Survivied','Not Survived')
bar_mf.plot.bar(color='Red')
```
